```python
#renaming columns to have better access to them (creating dictionary)

new_cols_dict ={
    'timestamp':'time',
    'cnt':'count',
    't1':'real_temp_C',
    't2':'feel_temp_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

#renaming columns in dataframe
df.rename(new_cols_dict, axis=1, inplace=True)
```

```python
df
```

```python
#changing humidity value to percentage (0 - 1)
df.humidity_percent = df.humidity_percent / 100
```

```python
df.humidity_percent
```

```python
#changing seasons codes to seasons names / changing float to written values (creating dictionary)
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}

#changing weather codes to weather names / changing float to written values (creating dictionary) 
#1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity 
#2 = scattered clouds /few clouds 
#3 = Broken clouds 
#4 = Cloudy 
#7 = Rain/ light Rain shower/ Light rain 
#10 = rain with thunderstorm 
#26 = snowfall 
#94 = Freezing Fog

weather_dict = {
    '1.0':'clear',
    '2.0':'scattered clouds',
    '3.0':'broken clouds',
    '4.0':'cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall',
    '94.0':'Freezing Fog',
}

#changing data type to string
df.season = df.season.astype('str')
#mapping values to dictionary ones
df.season = df.season.map(season_dict)

#changing data type to string
df.weather = df.weather.astype('str')
#mapping values to dictionary ones
df.weather = df.weather.map(weather_dict)
```
